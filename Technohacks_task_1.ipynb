## Removing outlier using IQR method

```python
def remove_outliers_IQR(data,age,th=1.5):
    Q1=np.quantile(data["Age"],0.25)
    Q3=np.quantile(data["Age"],0.75)
    IQR=Q3-Q1
    lower_bound=Q1-th*IQR
    upper_bond=Q3+th*IQR
    return(data[(data["Age"]>=lower_bound) & (data["Age"]<=upper_bond)])
```

```python
column_to_check=["Age"]
for col in column_to_check:
    data2=remove_outliers_IQR(data,col)
```

```python
plt.figure(figsize=(10,4))
plt.boxplot([data['Age'],data2['Age']],labels=["with_outlier","without_outlier"])
plt.title("Comparison of Age with and without outlier")
plt.xlabel("Data with and without outlier")
plt.ylabel("Age values")
plt.show()
```
